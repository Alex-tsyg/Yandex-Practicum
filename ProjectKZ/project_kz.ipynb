```python
def engine_cleaner(cell):
    'функция для удаления лишних символов из столбца объем двигателя'
    good_symbols = ['1', '2', '3', '4', '5', '6', '7', '8', '9', '0', '.',] # список символов, которые мы оставляем
    new_cell = '' # новая пустая строка куда мы будем добавлять "хорошие" символы
    for element in cell:
        if element in good_symbols:
            new_cell += element
    if new_cell.endswith('.'): # отдельно обработаем момент где в конце остается точка
        new_cell = new_cell[:-1]
    return new_cell
```

```python
# для корректной работы функции приведем столбец к типу строка
df.engine_volume = df.engine_volume.astype('str')
df.engine_volume = df.engine_volume.str.replace(',', '.')
# применяем функцию
df.engine_volume = df.engine_volume.apply(engine_cleaner)
```

Ярко выраженный один объем двигателя есть у 5 моделей. Заполняем наиболее частым значением и заменим тип данных на float

Часть данных заполним значениями, взятыми из интернет, это этого напишем функцию замены значений и применим ее к столбцу.

```python
def engine(cell):
    if cell in['Nexia R3', 'R3']:
        return 1.5
    elif cell in['Niva']:
        return 1.7
    elif cell in['S5']:
        return 2.0
    elif cell in['S3']:
        return 1.6
    elif cell in['2310','2705','2752']:
        return 2.5
    elif cell in['3302','Next','3221']:
        return 2.7
    elif cell in['3308']:
        return 4.67
    elif cell in['6606','66052','2752']:
        return 6.7
    elif cell in['4311,''43118','65225','45141','58815Z','BJ3253DMPKB-AD']:
        return 11
    else:
        return cell
```

```python
# применим функцию
df['engine_volume']=df['engine_volume'].apply(engine)
# исправим ошибки для Duster
df['engine_volume'].loc[(df['model'] == 'Duster')&(df.fuel=='2')] = 2
df['engine_volume'].loc[(df['model'] == 'Duster')&(df.fuel=='1.6')] = 1.6
# заполним None заглушкой 0
df.engine_volume = df.engine_volume.replace('None', 0)
```

```python
# Функция замены значений
# Двигатель бооее 13л считаем ошибкой ввода, в файле нет моделей с объемом более 13
def replace_liter(value):
    if  value >=13 :
        return value/10
    elif value > 10 and value < 13 :
        return 10.0
    elif value >1.9 and value < 2.01:
        return 2.0
    else:
        return round(value,1)
```

```python
#применим функцию
df['engine_volume']=df['engine_volume'].apply(replace_liter)
```

### Проверить гипотезу: стоимость автомобиля зависит от уровня развития региона (средней заработной платы по региону);

Проверка гипотезы: стоимость автомобиля зависит от уровня развития региона<br>
стоимость автомобиля определим как среднюю по региону    <br>

<b>H_0: стоимость автомобиля зависит от уровня развития региона</b><br>
<b>H_a: стоимость автомобиля не зависит от уровня развития региона</b><br>
<b>alpha = 0.05</b><br>

Возьмем из интернета среднюю зп по региону и посчитаем среднюю стоимость авто по региону<br>
Область Экспорт отфильтруем<br>

[Средняя зп по Казахстану](https://stat.gov.kz/ru/industries/labor-and-income/stat-wags/)

```python
data = {'oblas': ['г.Алматы', 'г.Нур-Султан', 'Восточно-Казахстанская область',
       'Западно-Казахстанская область', 'Костанайская область',
       'Павлодарская область', 'Южно-Казахстанская область',
       'Мангистауская область', 'Атырауская область',
       'Актюбинская область', 'Карагандинская область',
       'Северо-Казахстанская область', 'Акмолинская область',
       'Кызылординская область', 'Жамбылская область',
       'Туркестанская область', 'Алматинская область'],
       'salary': [224158, 226796, 162182, 183914,145890,160670,110254,294099,
                  351103,156595,172239,130233,140272,152085,127043,123853,136212]}

salary2 = pd.DataFrame(data).sort_values('oblas',ascending=False)
reg_sr= df.groupby('oblas', as_index=False).agg({'price_usd':'mean'}).sort_values(by='oblas', ascending=False)
reg_sr.columns=['oblas', 'mean']
reg_sr = reg_sr[reg_sr['oblas'] != 'Экспорт область']
reg_sr['mean']= round(reg_sr['mean'],0).astype(int)
for_test=reg_sr.merge(salary2, on="oblas")
display(for_test)
x=for_test['mean']
y=for_test.salary
#p-value
alpha = 0.05 #задаем значение уровня значимости
p = round(pearsonr(x, y)[1], 4)
# вывод значения p-value на экран
print('p-value=', p)
# условный оператор с выводом строки с ответом
if p < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
bcg_df= mercur.groupby(['model','month_num'])['total_usd'].sum().reset_index()

bcg_df['perc_total']=bcg_df.groupby('model')['total_usd'].transform('sum')/bcg_df.total_usd.sum()*100
bcg_df['total']=bcg_df.groupby('model')['total_usd'].transform('sum')

bcg_df['grow']=bcg_df.groupby('model')['total_usd'].transform('last')/bcg_df.sort_values(by='month_num').groupby('model')['total_usd'].transform('first')*100
#

display(bcg_df)
# цвет кругов в зависимости от квадрата
color_scheme=['#0492c2']*11
color_scheme[5]='#028A0F'
color_scheme[0],color_scheme[2]='#FCAE1E','FCAR1E'
fig = px.scatter(bcg_df[bcg_df.perc_total>1.2],
                 x='perc_total', y='grow',
                 size='total', color='model',
                 size_max=75,
                 text='model',
                 #color_discrete_sequence=color_scheme,
                 title='BCG matrix Mercur Auto',
                 width=1000, height=600)
#квадраты
fig.add_shape(type='line',x0=0, y0=225,x1=40,y1=225, line=dict(color='black',dash='dash',width=3))
fig.add_shape(type='line',x0=20, y0=0,x1=20,y1=500, line=dict(color='black',dash='dash',width=3))
#подписи
fig.add_annotation(x=18.5, y=200, text='<b>Собаки</b>', showarrow=False)
fig.add_annotation(x=16.5, y=490, text='<b>Проблемные дети</b>', showarrow=False)
fig.add_annotation(x=38.5, y=490, text='<b>Звезды</b>', showarrow=False)
fig.add_annotation(x=37  , y=200, text='<b>Дойные короы</b>', showarrow=False)
#уберем легенду
fig.update_layout(showlegend=False, xaxis_title='Доля прибыли в%', yaxis_title='Изменение объема продаж %')
fig.show()
```
