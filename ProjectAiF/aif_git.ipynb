### Создаем функцию анализа датасета

```python
def show_stats(df: pd.core.frame.DataFrame) -> pd.core.frame.DataFrame:
    '''
    Функция для отображения информации о датасете
    '''
    start_shape = df.shape
    mins, means, medians, maxs = [], [], [], []
    stats_df = pd.DataFrame(index=df.columns.tolist())
    stats_df['dtype'] = [df[col].dtypes for col in df.columns]
    stats_df['nans_count'] = df.isna().sum().tolist()
    stats_df['nans_ratio'] = (stats_df.nans_count / start_shape[0] * 100).round(2)
    stats_df['duplicates_count'] = df.duplicated().sum().tolist()
    stats_df['duplicates_ratio'] = (stats_df.duplicates_count / start_shape[0] * 100).round(2)
    stats_df['uniques'] = [df[col].nunique() for col in df.columns]
    numeric_columns = ['nans_count', 'nans_ratio', 'uniques', 'min', 'mean', 'max']
    for col in df.columns:
        if df[col].dtypes in ['int', 'float']:
            mins.append(round(df[col].min(), 2))
            means.append(round(df[col].mean(), 2))
            #medians.append(round(df[col].median(), 2))
            maxs.append(round(df[col].max(), 2))
        else:
            mins.append(0)
            means.append(0)
            #medians.append(0)
            maxs.append(0)
    stats_df['min'], stats_df['mean'],  stats_df['max'] = mins, means,  maxs
    display(stats_df)
```

Создадим вспомогательную функцию разбивки на сегменты

```python
def make_segments(cell: str) -> str:
    'присвоение типа клиенту в зависимости от RFM сегмента'
    #new_cell = ''
    
    if cell =='000':
        new_cell = ''        
    elif cell in ['333']:
        new_cell = 'ключевой'    
    elif cell in ['332', '331', '322']:
        new_cell = 'лояльный'
    elif cell in ['323', '233', '223']:
        new_cell = 'крупный'
    elif cell in ['232', '231', '222', '221', '212', '211']:
        new_cell = 'неактивный'
    elif cell in ['213', '133', '123', '113']:
        new_cell = 'крупный неактивный'
    elif cell in ['321', '313', '312', '311']:
        new_cell = 'новый'
    elif cell in ['132', '131', '122', '121']:
        new_cell = 'почти потерянный'
    else:
        new_cell = 'ушедший'
    return new_cell
```

<font face='Liberation serif' size=4 color='blue'>*RFM-анализ*</font>

RFM-анализ — это анализ клиентской базы на основе истории покупок, проще говоря, сегментация по частоте, давности и сумме трат.
* давность (Recency) — как давно совершил последний заказ
* частота (Frequency) — как часто совершает заказы
* деньги (Monetary) — сколько потратил на услуги<br>
<br>
RFM-анализ базируется на законе Парето: 20% усилий дают 80% результата, если перевести на язык продаж, то 20% всей клиентской базы делают для компании 80% продаж. Наша цель — выделить эти 20% самых важных покупателей и направить максимум усилий на их удержание, а также определить бесперспективных, чтобы не тратить на них много ресурсов.

### RFM анализ: сегменты, статистика по сегментам, поведение по сегментам (сколько “живут” люди в разных сегментах, как часто и сколько тратят на благотворительность, какой канал привлечения, на какие активности тратили деньги

Для анализа нам нужны данные лишь об успешнхя платежах, поэтому отфильтуем датасет заказов по этому признаку

```python
order_flt=order[order.status=='Paid']
```

Дату анализа выставим максимальную дату в датасете + 1 день

```python
work_date = order_flt.date_pay.max() + pd.Timedelta('1d')  
```

```python
work_date
```

Создадим RFM таблицу: тут понадобится количество заказов на клиента, сумма покупок и даты жизни клиента: минимальная и максимальная даты

```python
rfm = order_flt.groupby('customer_id').agg({'pay_id': 'count',
                                     'total': 'sum',
                                     'date_pay': ['min', 'max']}).reset_index()
```

Переименуем колонки 

```python
rfm.columns = ['customer_id', 'orders_count', 'total_sales', 'start_date', 'end_date']
```

Период жизни клиента - разница между конечной и начальной датой

```python
rfm['period'] = (rfm.end_date - rfm.start_date) / pd.Timedelta('1d') + 1
```

recency - это время с момента последнего заказа: наша дата анализа (work date) минус дата когда клиент последний раз делал заказ

```python
rfm['r_value'] = (work_date - rfm.end_date) / pd.Timedelta('1d')
```

Частоту лучше привести к одной шкале - для этого мы количество заказов делим на период жизни клиента. 

```python
rfm['f_value'] = (rfm.orders_count / rfm.period).round(2)
```

Monetary - это сумма всех заказов клиента, эта информация у нас уже есть в столбце total sales

```python
rfm['m_value'] = rfm.total_sales
```
